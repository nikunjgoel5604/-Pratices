4. An armstrong number is the one whose sum of nth power of its digits is equal to the number itself ypur task is yo print all the arm strong number in the range (lower , upper) of no arm strong number is found b/w the given range print 1

```python
# Solution 4
# Prints Armstrong Numbers in given range
def findArmstrong(low, high):

    for i in range(low + 1, high):

        # number of digits calculation
        x = i
        n = 0
        while x != 0:
            x = x // 10          # use integer division
            n = n + 1

        # compute sum of nth power of digits
        pow_sum = 0
        x = i
        while x != 0:
            digit = x % 10
            pow_sum = pow_sum + digit ** n   # use ** instead of math.pow()
            x = x // 10                      # use integer division

        # check if number is Armstrong
        if pow_sum == i:
            print(i, end=" ")

# Driver code
num1 = 1000
num2 = 10000
findArmstrong(num1, num2)
print()
```

Ques 5: Given a postive numver x. check the whether number x is a perfect number or not. A perfect number is a number which is equal to the sum of its proper positive divisors.

```python
# Function to check if the number is perfect and show its divisors and sum
def isPerfect(n):
    total_sum = 0
    divisors = []  # to store all divisors

    for i in range(1, n):
        if n % i == 0:              # check if i is a divisor of n
            divisors.append(i)     # store the divisor
            total_sum += i         # add divisor to sum

    print("\nAll divisors are:", divisors)
    print("Sum of divisors:", total_sum)

    return total_sum == n          # return True if perfect number


if __name__ == '__main__':
    n = 28
    print("\n Perfect Number:", "true" if isPerfect(n) else "false")
```
