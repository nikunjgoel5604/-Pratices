Merge Sort 

you are given two sorted lists of size m and n. Merge them in a given  sorted array in sorted oreder and print the sum of two middle element after merging the array .


```python
def sum_of_middle_element(a, b):
    # define (def) a function to merge two sorted arrays and find sum of middle elements

    m, n = len(a), len(b)
    # find the length of array one and two (a, b)

    i = j = 0
    # initialize the pointer for both the arrays to know which element we have to compare next

    merged = []
    # creating an empty array to store the merged sorted array

    while i < m and j < n:
        # merge both arrays while elements are left in both array a and b

        if a[i] < b[j]:
            # if the current element of a is smaller then add it to merged list
            merged.append(a[i])
            i += 1
            # move the pointer of array a to the next element

        else:
            # otherwise add current element of b to merged list
            merged.append(b[j])
            j += 1
            # move the pointer of array b to the next element

    while i < m:
        # if any elements are left in array a, add them to merged list
        merged.append(a[i])
        i += 1
        # move the pointer of array a to the next element

    while j < n:
        # if any elements are left in array b, add them to merged list
        merged.append(b[j])
        j += 1
        # move the pointer of array b to the next element

    total = m + n
    # total number of elements in merged list

    if total % 2 == 0:
        # if the total number of elements is even

        mid1 = merged[total // 2 - 1]
        # first middle element (left middle)

        mid2 = merged[total // 2]
        # second middle element (right middle)

        print("Middle elements are:", mid1, "and", mid2)
        # printing the middle elements

        return mid1 + mid2
        # return the sum of both middle elements

    else:
        # if total number of elements is odd

        mid = merged[total // 2]
        # take the single middle element

        print("Middle element is:", mid)
        # printing the middle element

        return mid * 2
        # return its double (same element counted twice)


# Taking input for first sorted list
A = list(map(int, input("Enter elements of first sorted list: ").split()))
# input multiple elements for array A in one line

# Taking input for second sorted list
B = list(map(int, input("Enter elements of second sorted list: ").split()))
# input multiple elements for array B in one line

print("Array A:", A)
# print the first array

print("Array B:", B)
# print the second array

result = sum_of_middle_element(A, B)
# calling the function and storing the returned sum

print("Sum of middle elements:", result)
# printing the final answer
```
