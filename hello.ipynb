##### PLoting 


```python
#1. Matplotlib: Bar chart for Total Sales by Category
import pandas as pd
import matplotlib.pyplot as plt

# Load the Global Superstore Data into a pandas DataFrame
df = pd.read_csv('Data Set/superstore.csv')

# Group data by Category and sum the Sales to get total sales per category
category_sales = df.groupby('Category')['Sales'].sum()

# Create a bar chart to visualize the total sales per category
plt.bar(category_sales.index, category_sales.values, color=['skyblue', 'orange', 'green'])

# Set the title and labels for the chart
plt.title('Total Sales by Category')  # Title of the chart
plt.xlabel('Category')  # Label for the x-axis
plt.ylabel('Sales')  # Label for the y-axis

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the chart
plt.show()


```

```python
#2.Seaborn: Box plot for Profit by Sub-Category
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = "Data Set/superstore.csv"
df= pd.read_csv(df)

# Select the relevant columns (Sub.Category and Profit) for visualization
subcat_profit = df[['City', 'Order.ID']]

# Set the figure size for the plot
plt.figure(figsize=(52, 48))

# Create a box plot to visualize the profit distribution for each sub-category
sns.boxplot(x='City', y='Order.ID', data=subcat_profit, )

# Set the title and labels for the chart
plt.title('Profit Distribution by Sub-Category')  # Title of the chart
plt.xticks(rotation=360)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

```python
#3. Plotly: Interactive Pie chart for Sales by Region
import plotly.express as px

# Group data by Region and calculate the total sales for each region
region_sales = df.groupby('Region')['Sales'].sum().reset_index()

# Create an interactive pie chart to visualize sales distribution across regions
fig = px.pie(region_sales, names='Region', values='Sales', title='Sales by Region', 
             color_discrete_sequence=px.colors.sequential.RdBu)

# Display the interactive pie chart
fig.show()
```
